```python
import pandas as pd
import csv
import numpy as np
```

```python
data = pd.read_csv('mobile_price_train.csv')
data.head(6)
```

```python
data['n_cores'].value_counts()
```

```python
#calculating p-value for battery_power:

#population mean:
population_mean = data['battery_power'].mean()
print("population mean for battery power: ",population_mean)

#sample mean:
sample = 200
sum = 0.0
for i in range(sample):
    sum += data['battery_power'].iloc[i]
    sample_mean = sum/sample
print("Sample mean of battery power: ",sample_mean)

#calculating sample standard deviation:
deviation_sum = 0
for i in range(sample):
    deviation_sum += abs((data['battery_power'].iloc[i]- sample_mean))**2   
sample_std = np.sqrt(deviation_sum/sample)
print("sample std for battery power: ",sample_std)

#calculating z-score:
z = (sample_mean - population_mean)/(sample_std/np.sqrt(sample))
print("z-score for battery power: ",z)

#calculating p-value:
p_value = 2*(0.36) #probability for z-score(-0.37) = 0.36
print("p-value for battery power: ",p_value)

#suppose significance level alpha = 0.05
if p_value > 0.05:
    print("Feature is statistically significant")
else:
    print("Feature is not statistically significant")
```
