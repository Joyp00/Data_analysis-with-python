```python
import pandas as pd
import csv
import numpy as np
```

```python
data = pd.read_csv('great_customers.csv')
data.head()
```

```python
#cleaning data:
data['mins_exercising_year'].fillna(0,inplace=True)

#Hypothesis testing for totChol:
pop_mean = data['mins_exercising_year'].mean()
print("population mean: ",pop_mean)
pop_std = data['mins_exercising_year'].std()
print("population standard deviation: ",pop_std)

#Suppose null hypothesis = 44
null_hypothesis = 44
sample = 200

#sample mean:
sample = 200
sum = 0.0
for i in range(sample):
    sum += data['mins_exercising_year'].iloc[i]
    sample_mean = sum/sample
print("Sample mean: ",sample_mean)


#calculating z-score:
z = (sample_mean - null_hypothesis)/(pop_std/np.sqrt(sample))
print("z-score: ",z)

#critical value = 1.96
if z > 1.96:
    print("Reject null hypothesis")
elif z < -1.96:
    print("Reject null hypothesis")
else:
    print("Do no reject null hypothesis")

```
