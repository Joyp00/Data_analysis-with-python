```python
import pandas as pd
import csv
import numpy as np
```

```python
data = pd.read_csv('heart_disease.csv')
data.head()
```

```python
data['glucose'].value_counts()
```

```python
v = 0.0
def p_value(v):
    if v >= .5 and v <= .99997:
        print("p_value: ",2*(1-v))
        if 2*(1-v) > 0.05:
            print("Feature is not rejected")
        else:
            print("Feature is rejected")
       
    elif v <= .46414 and v >= .00005:
        print("p_value: ",abs(2*v))
        if abs(2*v) > 0.05:
            print("Feature is not rejected")
        else:
            print("Feature is rejected")    
```

```python
#calculating p-value for male:

#population mean:
population_mean = data['male'].mean()
print("population mean for male: ",population_mean)

#sample mean:
sample = 200
sum = 0.0
for i in range(sample):
    sum += data['male'].iloc[i]
    sample_mean = sum/sample
print("Sample mean of male: ",sample_mean)

#calculating sample standard deviation:
deviation_sum = 0
for i in range(sample):
    deviation_sum += abs((data['male'].iloc[i]- sample_mean))**2   
sample_std = np.sqrt(deviation_sum/sample)
print("sample std for male: ",sample_std)

#calculating z-score:
z = (sample_mean - population_mean)/(sample_std/np.sqrt(sample))
print("z-score for male: ",z)
if z > 3.9 or z < -3.9:
    print("Feature is rejected")
else:
    #calculating p-value:
    p_value(0.16)            #probability for z-score(-0.99) = 0.16 #suppose significance level alpha = 0.05
    
```
